```python
import sys
import scipy.io.wavfile
import numpy as np
import matplotlib.pyplot as plt
 
 
#音声ファイル読み込み
args = sys.argv
wav_filename = '../wav/17710517.wav'
rate, data = scipy.io.wavfile.read(wav_filename)
length = data.shape[0] / rate

print(rate)
print(data.shape[0])
print(length)
print(length * 29.7)
 
data = data / 32768
time = np.arange(0, length, 1/rate)  

plt.plot(time, data)
plt.xlim(0, 1000/rate)
plt.show()
```

```python
K = 256
for i in range(10):
    plt.figure(figsize=(16, 2))
    plt.plot(time[(K*i):(K*(i+1))], data[(K*i):(K*(i+1))])
    plt.show()
```
